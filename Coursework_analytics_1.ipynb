# 1.2. Определим количество модулей на каждом курсе:

```python
# Сгруппируем датасет 'work_df' по столбцу 'course_id' и получим длины списков уникальных номеров модулей (то есть
# количество модулей), преобразуя всё в датасет:

grouped_df_1_2 = work_df.groupby('course_name')['module_number'].unique().apply(lambda x : len(x)).to_frame().reset_index()

# Переименуем столбец 'module_number', чтобы заголовок отражал суть содержимого столбца:

grouped_df_1_2.rename(columns = {'module_number':'count_of_modules'}, inplace = True)
```

# 1.3. Определим количество уроков в каждом модуле на каждом курсе:

```python
# Сгруппируем датасет 'work_df' по столбцам 'module_number' и  'module_number' и получим длины списков уникальных 
# номеров уроков (то есть количество уроков), преобразуя всё в датасет:

grouped_df_1_3 = work_df.groupby(['course_name', 
                                  'module_number'])['lesson_number'].unique().apply(lambda x : len(x)).to_frame().reset_index()

# Переименуем столбец 'lesson_number', чтобы заголовок отражал суть содержимого столбца:

grouped_df_1_3.rename(columns = {'lesson_number':'count_of_lesson'}, inplace = True)
```

```python
# Получим датасет из датасета из пункта 1.3, сбросив индексы, сгруппировав по индексу курса, и посчитав медианное значение
# количества уроков на каждом курсе:

grouped_df_1_4 = grouped_df_1_3.reset_index().groupby(['course_name'])['count_of_lesson'].median().to_frame().reset_index()

# Переименуем столбец 'count_of_lesson', чтобы заголовок отражал суть содержимого столбца:

grouped_df_1_4.rename(columns = {'count_of_lesson':'median_count_of_lesson'}, inplace = True)
```

# 1.6. Определим минимальный, максимальный, средний и медианный возраст студентов:

```python
# Создаём датафрейм на основе csv-файла:

students_df = pd.read_csv('students.csv')

# Уберём из датасета столбец с местом жительства, для удобства:

students_df_2 = students_df.drop(['city'], axis = 1)

# Уберём строки с пропущенными значениями в столбце 'birthday':

students_df_2.dropna(axis = 0, how ='any', inplace = True)

# Создадим столбец 'real_birthday', в котором будут только корректные значения года рождения или значения Nan:

students_df_2['real_birthday'] = students_df_2['birthday'].apply(lambda x: x if 1900 < int(x[:4]) < 2010 else np.NaN)

# Снова уберём строки с пропущенными значениями в столбце:

students_df_2.dropna(axis = 0, how ='any', inplace = True)

# Обновим столбец 'birthday', записав в него даты рождения в формате datetime:

students_df_2['birthday'] = pd.to_datetime(students_df_2['real_birthday'], format='%Y-%m-%d')

# Удалим вспомогательный столбец:

students_df_2.drop(['real_birthday'], axis = 1, inplace = True)

# Определим возраст самого возрастного студента:

old_age = 2021 - students_df_2['birthday'].dt.year.min()

# Определим возраст самого молодого студента:

young_age = 2021 - students_df_2['birthday'].dt.year.max()

# Определим медианный возраст студентов:

median_age = 2021 - students_df_2['birthday'].dt.year.median()
```

# Возраст самого возрастного студента:

```python
old_age
```

# Возраст самого молодого студента:

```python
young_age
```

# Медианный возраст студентов:

```python
median_age
```

# 1.7. Определим минимальный, максимальный, средний и медианный возраст студентов на каждом курсе:

```python
# Создадим сводную таблицу:

students_df_3 = pd.merge(work_df, students_df_2, how='left', left_on='student_id', right_on='id')

# Удалим столбец 'finish_date', чтобы не терять строки, в которых finish_date is Nan:

students_df_3.drop(['finish_date'], axis = 1, inplace = True)

# Уберём строки с пропущенными значениями:

students_df_3.dropna(axis = 0, how ='any', inplace = True)

# Получим новый датасет, произведя группировку по индексу курса и id студента, и получим столбец с датами рождения, 
# соответствующими студентам, при этом сбросив индексы:

students_df_4 = students_df_3.groupby(['course_name', 'student_id'])['birthday'].first().to_frame().reset_index()

# Добавим столбец с возрастом студентов:

students_df_4['age'] = 2021 - students_df_4['birthday'].dt.year
```

# Возраст самого молодого студента на каждом курсе:

```python
students_df_4.groupby(['course_name'])['age'].min().to_frame()
```

# Возраст самого возрастного студента на каждом курсе:

```python
students_df_4.groupby(['course_name'])['age'].max().to_frame()
```

# Средний возраст студентов на каждом курсе:

```python
students_df_4.groupby('course_name')['age'].mean().to_frame()
```

# Медианный возраст студентов на каждом курсе:

```python
students_df_4.groupby('course_name')['age'].median().to_frame()
```

```python
# Создадим новый датафрейм на основе 'work_df', копируя при этом только строки, содержащие домашние задания:

work_df_2_1 = work_df[work_df['is_homework'] == True][['start_date', 'course_id', 'course_name', 'student_id']].copy()

# Преобразуем тип столбца 'start_date' в datetime:

work_df_2_1['start_date'] = pd.to_datetime(work_df_2_1['start_date'], format='%Y-%m-%d')

# Добавим столбец с годом по дате:

work_df_2_1['year'] = work_df_2_1['start_date'].dt.year

# Добавим столбец с месяцем по дате:

work_df_2_1['month'] = work_df_2_1['start_date'].dt.month

# Сгруппируем датасет 'work_df_2_1' по столбцам 'course_id', 'year', 'month', 'student_id', 
# произведя при этом сортировку по датам:

grouped_df_2_1 = work_df_2_1.groupby(['course_id', 'year', 'month', 'student_id']).agg({'course_name' : 'first'}).reset_index()

# Сгруппируем датасет 'grouped_df_2_1' по столбцам 'course_id' и 'student_id', применим агрегирующие функции 'first',
# чтобы получить первые упоминания о конкретных студентах на каждом курсе:

grouped_df_2_1_2 = grouped_df_2_1.groupby(['course_id', 'student_id'], sort=False).agg({'year' : 'first',
                                                                             'month' : 'first',
                                                                             'course_name' : 'first'}).reset_index()

# Сгруппируем датасет 'grouped_df_2_1_2' по столбцам 'course_id', 'year', 'month', применим агрегирующую функцию 'count',
# чтобы получить количество новых студентов в каждый контрольный месяц каждого года:

grouped_df_2_1_3 = grouped_df_2_1_2.groupby(['course_id', 'year', 'month']).agg({'student_id' : 'count',
                                                                             'course_name' : 'first'}).reset_index()

# Добавим в датафрейм новый столбец с датой замера прироста учеников: 

grouped_df_2_1_3['control_time'] = pd.to_datetime(grouped_df_2_1_3.year.astype(str) + grouped_df_2_1_3.month.astype(str), format='%Y%m')

# Переименуем столбец 'student_id', чтобы заголовок отражал суть содержимого столбца:

grouped_df_2_1_3.rename(columns = {'student_id':'increment'}, inplace = True)
```

# 4.1. Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе.
Формула такая: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
# Создадим новый датафрейм на основе 'work_df':

work_df_4_1 = work_df[work_df['is_homework'] == True].copy()

# Произведём группировку датафрейма "work_df_4_1" по имени курса, номеру модуля, id студента и статусу:

grouped_df_4_1 = work_df_4_1.groupby(['course_name',
                                      'field',
                                      'module_number', 
                                      'module_title', 
                                      'student_id', 
                                      'status']).agg({'lesson_number' : 'max',}).reset_index()

# Произведём группировку датафрейма "grouped_df_4_1" по имени курса, номеру модуля и статусу, чтобы
# посчитать количество студентов для каждого статуса выполнения:

grouped_df_4_1_2 = grouped_df_4_1.groupby(['course_name',
                                           'field',
                                           'module_number', 
                                           'module_title',
                                           'status']).agg({'student_id' : 'count'}).reset_index()

# Переименуем столбец 'student_id':

grouped_df_4_1_2.rename(columns={'student_id':'count_of_students'}, inplace=True)

# Создадим столбец для добавления в него только выполнивших ДЗ студентов:

grouped_df_4_1_2['count_of_done_students'] = grouped_df_4_1_2.apply(lambda x: x['count_of_students'] 
                                                     if (x['status'] == 'done') else 0, axis=1)

# Сгруппируем датафрейм 'grouped_df_4_1_2' по имени курса и номеру модуля, чтобы посчитать общее количество студентов и
# количество студентов, приступивших к заданию:

grouped_df_4_1_3 = grouped_df_4_1_2.groupby(['course_name',
                                             'field',
                                             'module_number', 
                                             'module_title']).agg({'count_of_students' : 'sum',
                                                                   'count_of_done_students' : 'sum'}).reset_index()

# Заменим нулевые значения в столбце 'count_of_done_students' на 1, чтобы дальше не иметь проблем с делением на 0:

grouped_df_4_1_3.loc[(grouped_df_4_1_3['count_of_done_students'] == 0), 'count_of_done_students'] = 1

# Создадим список для записи в него значений конверсии:

conversion_list = []

# Посчитаем конверсию перехода студентов из одного модуля в другой на каждом курсе:

for count, item in enumerate(grouped_df_4_1_3['count_of_students']):
    
    if grouped_df_4_1_3['module_number'][count] == 1:
    
        conversion_list.append(1)
        
    else:
            
        conversion_list.append((grouped_df_4_1_3['count_of_students'][count])/
                                (grouped_df_4_1_3['count_of_done_students'][count - 1]))
        
# Создадим и заполним столбец со значениями конверсий:

grouped_df_4_1_3['conversion'] = conversion_list

# Создадим и заполним столбец, содержащий Номера и наименования модулей:

grouped_df_4_1_3['module_title_number'] = grouped_df_4_1_3.apply(lambda x: (str(x['module_number']) + ': ' 
                                                                 + str(x['module_title']))[:55] + '..'
                                                                 if len(str(x['module_title'])) > 55
                                                                 else (str(x['module_number']) + ': ' 
                                                                       + str(x['module_title'])), axis=1)
```
